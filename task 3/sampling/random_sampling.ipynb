```python
import random
import pandas as pd
```

```python
#read dataset
authors=pd.read_csv('author_result.csv')
authors
```

```python
#Title
Title=list(authors['Title'])
Title=random.choices(Title,k=214)
Title
```

```python
# Citation
Citation=list(authors['Citation'])
Citation=random.choices(Citation,k=214)
# Citation
```

```python
# DOI
DOI=list(authors['DOI'])
DOI=random.choices(DOI,k=214)
# DOI
```

```python
# Year
Year=list(authors['Year'])
Year=random.choices(Year,k=214)
# Year
```

```python
# FINDINGS
FINDINGS=list(authors['FINDINGS'])
FINDINGS=random.choices(FINDINGS,k=214)
# FINDINGS
```

```python
#institution names
aff=list(authors['affiliation'])
affiliation=random.choices(aff,k=214)
# affiliation
```

```python
# lab_size
lab_size=list(authors['lab_size'])
lab_size=random.choices(lab_size,k=214)
# lab_size
```

```python
# Highest_degree
Highest_degree=list(authors['Highest_degree'])
Highest_degree=random.choices(Highest_degree,k=214)
# Highest_degree
```

```python
# Degree_area
Degree_area=list(authors['Degree_area'])
Degree_area=random.choices(Degree_area,k=214)
# Degree_area
```

```python
# Publication Rate
Publication_Rate=list(authors['Publication Rate'])
Publication_Rate=random.choices(Publication_Rate,k=214)
# Publication_Rate
```

```python
# Other Journals Published In
Other_Journals=list(authors['Other Journals Published In'])
Other_Journals=random.choices(Other_Journals,k=214)
# Other_Journals
```

```python
# duration of Career(Years)
duration=list(authors['duration of Career(Years)'])
duration=random.choices(duration,k=214)
# duration
```

```python
data={
    'Title':Title,
    'Citation':Citation,
    'DOI':DOI,
    'Year':Year,
    'FINDINGS':FINDINGS,
    'affiliation':affiliation,
    'lab_size':lab_size,
    'Highest_degree':Highest_degree,
    'Degree_area':Degree_area,
    'Publication_Rate':Publication_Rate,
    'Other Journals Published In':Other_Journals,
    'duration of Career(Years)':duration   
}
df=pd.DataFrame(data)
df
```

```python
df.to_csv('sampling.csv')
```
