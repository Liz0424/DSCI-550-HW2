```python
import requests
import pandas as pd
```

```python
#read dataset
f=open("names.txt")
line = f.readline()
columns=[]
while line:
    columns.append(line.rstrip('\n'))
    line = f.readline()
columns
```

```python
#create names
first_names=[]
last_names=[]
for i in range(len(columns)):
    if i%2==0:
        first_names.append(columns[i])
    else:
        last_names.append(columns[i])
print(first_names)
print(last_names)
        
```

```python
data={
    'first_name':first_names,
    'last_name':last_names
}
df=pd.DataFrame(data)
df
```

```python
df['full_name']=df['first_name']+' '+df['last_name']
df
```

```python
df.to_csv('author_name.csv')
```
