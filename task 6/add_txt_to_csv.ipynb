```python
import pandas as pd
read_csv=pd.read_csv('task_6_info.csv')
```

```python
read_csv.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
read_csv
```

```python
import os,glob
import os.path

ext = "*.txt"
PATH = "generation_texts"

# Find all the files with that extension
files = []
for dirpath, dirnames, filenames in os.walk(PATH):
    files += glob.glob(os.path.join(dirpath, ext))
```

```python
files
```

```python
def read_txt(file):
    with open(file) as f:
        lines = f.readlines()
    return lines
```

```python
create_list=[]
for i in files:
    create_list.append(''.join(read_txt(i)))
```

```python
create_list[0]
```

```python
create_dict={}
create_dict['text']=create_list
```

```python
text_df=pd.DataFrame(create_dict)
text_df
```

```python
result = pd.concat([read_csv, text_df], axis=1, join='inner')
```

```python
result.to_csv('new_combine.csv')
```
