```python
import tika
from tika import parser
import os

# Specify Input and output directories
pdfs_directory_path = "pdfs"
output_directory_path = os.path.join(pdfs_directory_path,"text_outputs")


# Check if output directory exists, if not then create it
if not (os.path.exists(output_directory_path)):
    os.mkdir(output_directory_path)
```

## Use tika to extract texts from each file

```python

# Initialize tika
tika.initVM()

# Iterate over all pdf files, extracting text from each using tika
for filename in os.listdir(pdfs_directory_path):
    if filename.endswith(".pdf"):
        file_path_str  = os.path.join(pdfs_directory_path, filename)
        parsed = parser.from_file(file_path_str)
        print(parsed["metadata"])
        output_file_path = os.path.join(output_directory_path,filename.rstrip(".pdf") +".txt" )
        with open(output_file_path, 'w', encoding="utf8") as fo:
            fo.write(parsed["content"].strip())
```

## Clean & Combine the extracted text files
As the text extraction from PDFs included some artifacts such as due to specially formatted blocks in the PDF files, the below cleaning code removes those lines that have less than 5 characters in a line which is a common artifact seen in many of the text files extracted from the PDFs

```python
cleaned_lines=[]

for filename in os.listdir(output_directory_path):
    if filename.endswith(".txt"):
        file_path_str  = os.path.join(output_directory_path, filename)
        with open(file_path_str,'r', encoding="utf8") as fin:
            lines = fin.readlines()
        for line in lines:
            if len(line)>5:
                cleaned_lines.append(line)
                
with open('ALL_COMBINED_papers_CLEANED_for_GPT2_training.txt','w', encoding="utf8") as fo:
    fo.writelines(cleaned_lines)
```
